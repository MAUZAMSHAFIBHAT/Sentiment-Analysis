```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project2_Sentiment_Analysis/a1_RestaurantReviews_HistoricDump.tsv',delimiter = '\t', quoting = 3)
```

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0, 900):
  review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle
cvFile='drive/MyDrive/Colab Notebooks/Project2_Sentiment_Analysis/c1_BoW_Sentiment_Model.pkl'
cv = pickle.load(open(cvFile, "rb"))
```

```python
X_fresh = cv.transform(corpus).toarray()
X_fresh.shape
```

```python
import joblib
classifier = joblib.load('drive/MyDrive/Colab Notebooks/Project2_Sentiment_Analysis/c2_Classifier_Sentiment_Model')
```

```python
y_pred = classifier.predict(X_fresh)
print(y_pred)
```

```python
dataset['predicted_label'] = y_pred.tolist()
dataset.head()
```
